```python
import pandas as pd
```

```python
df = pd.read_csv('all_results/results_iterative_instances.csv')
df
```

```python
df.groupby(by=['model', 'prompt_type'])['energy'].mean()
```

```python
df = df[df['model'] == 'iterative_instances']
df = df[df['prompt_type'].str.contains('gemini_25_d')]
```

```python
df.sort_values(by=['energy']).head(3)
```

```python
for i, row in df.sort_values(by=['energy']).head(5).iterrows():
    print(row['dataset'], row['index'])
```

```python
df.groupby(by=['prompt_type', 'dataset'])['energy'].mean().sort_values(ascending=False)
```

```python
df.groupby(by=['model', 'prompt_type', 'dataset'])['energy'].mean()
```

```python
df['easy'] = (df['movement_dir'] == 'four') & (df['carry_limit'] == 100) & (df['cost_per_step'] == 0) & (df['dataset'] == 'inner_random_free')
df.groupby(by=['model', 'prompt_type', 'easy'])['energy'].mean()
```

```python
# Get average action sequence length
df.groupby(by=['model', 'prompt_type'])['actions'].apply(lambda x: x.str.split(' ').str.len().mean())
```
