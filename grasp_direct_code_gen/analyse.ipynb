```python
import pandas as pd
```

```python
baseline_df = pd.read_csv('all_results/results_baseline.csv')
baseline_df
```

```python
df = pd.read_csv('all_results/results_direct_code_gen.csv')
# df = pd.read_csv('all_results/results_greedy_extend_pseudocode.csv')
# df = pd.read_csv('all_results/results_step_by_step_code.csv')
df
```

```python
df = pd.concat([baseline_df, df])
df.reset_index(inplace=True, drop=True)
# df
```

```python
df.groupby(by=['model', 'prompt_type'])['energy'].mean().sort_values(ascending=False)
```

```python
df.groupby(by=['model', 'prompt_type', 'dataset'])['energy'].mean()
```

```python
df['easy'] = (df['movement_dir'] == 'four') & (df['carry_limit'] == 100) & (df['cost_per_step'] == 0) & (df['dataset'] == 'inner_random_free')
df.groupby(by=['model', 'prompt_type', 'easy'])['energy'].mean()
```

```python
# Average action sequence length
df.groupby(by=['model', 'prompt_type'])['actions'].apply(lambda x: x.str.split(' ').str.len().mean())
```
